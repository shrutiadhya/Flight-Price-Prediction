```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
train_df = pd.read_excel('Data_Train.xlsx')
test_df = pd.read_excel('Test_set.xlsx')
```

# Feature Engineering

```python
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.dtypes
```

```python
final_df=final_df.drop(['Date_of_Journey'],axis=1)
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].fillna('1 stop')
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].replace('non-stop','0 stop')
```

```python
final_df.head()
```

```python
final_df['Stops'] = final_df['Total_Stops'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
final_df['Stops'] = final_df['Stops'].astype(int)
final_df = final_df.drop(['Total_Stops'],axis=1)
```

```python
final_df.head()
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_Minute'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Minute'] = final_df['Arrival_Minute'].astype(int)
final_df = final_df.drop(['Arrival_Time'],axis=1)
```

```python
final_df.head()
```

```python
final_df['Dep_Hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_Minute'] = final_df['Dep_Time'].str.split(':').str[1]
final_df['Dep_Hour'] = final_df['Dep_Hour'].astype(int)
final_df['Dep_Minute'] = final_df['Dep_Minute'].astype(int)
final_df = final_df.drop(['Dep_Time'],axis=1)
```

```python
final_df.head()
```

```python
final_df['Route_1'] = final_df['Route'].str.split('→ ').str[0]
final_df['Route_2'] = final_df['Route'].str.split('→ ').str[1]
final_df['Route_3'] = final_df['Route'].str.split('→ ').str[2]
final_df['Route_4'] = final_df['Route'].str.split('→ ').str[3]
final_df['Route_5'] = final_df['Route'].str.split('→ ').str[4]
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df['Price'].fillna((final_df['Price'].mean()),inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df['Route_1'].fillna("None",inplace=True)
final_df['Route_2'].fillna("None",inplace=True)
final_df['Route_3'].fillna("None",inplace=True)
final_df['Route_4'].fillna("None",inplace=True)
final_df['Route_5'].fillna("None",inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df['Duration_Hour'] = final_df['Duration'].str.split('h').str[0]
final_df['Duration_Minute'] = final_df['Duration'].str.split(' ').str[1]
final_df['Duration_Minute'] = final_df['Duration_Minute'].str.split('m').str[0]
final_df['Duration_Minute'].fillna(0,inplace=True)
```

```python
final_df.head()
```

```python
final_df = final_df.drop(['Route'],axis=1)
final_df = final_df.drop(['Duration'],axis=1)
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
Airline_df = pd.get_dummies(final_df.Airline)
Source_df = pd.get_dummies(final_df.Source)
Destination_df = pd.get_dummies(final_df.Destination)
Additionalinfo_df = pd.get_dummies(final_df.Additional_Info)
Route1_df = pd.get_dummies(final_df.Route_1)
Route2_df = pd.get_dummies(final_df.Route_2)
Route3_df = pd.get_dummies(final_df.Route_3)
Route4_df = pd.get_dummies(final_df.Route_4)
Route5_df = pd.get_dummies(final_df.Route_5)
final_df = pd.concat([final_df, Airline_df,Source_df,Destination_df,Additionalinfo_df,Route1_df,Route2_df,Route3_df,Route4_df,Route5_df], axis=1)
```

```python
final_df=final_df.drop(['Airline'],axis=1)
final_df=final_df.drop(['Source'],axis=1)
final_df=final_df.drop(['Destination'],axis=1)
final_df=final_df.drop(['Additional_Info'],axis=1)
final_df=final_df.drop(['Route_1'],axis=1)
final_df=final_df.drop(['Route_2'],axis=1)
final_df=final_df.drop(['Route_3'],axis=1)
final_df=final_df.drop(['Route_4'],axis=1)
final_df=final_df.drop(['Route_5'],axis=1)
```

```python
pd.set_option('display.max_columns', None)
```

```python
final_df.loc[final_df['Duration_Hour']=='5m']
```

```python
final_df['Duration_Hour'] = final_df['Duration_Hour'].replace('5m','5')
```

# Feature Selection

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
final_df.shape
```

```python
df_train = final_df[0:10683]
df_test = final_df[10683:]
```

```python
X= df_train.drop(['Price'],axis=1)
y = df_train.Price
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
model = SelectFromModel(estimator=Lasso(alpha=0.005,random_state=0))
```

```python
model.fit(X_train,y_train)
```

```python
model.get_support()
```

```python
selected_features = X_train.columns[(model.get_support())]
```

```python
list(selected_features)
```

```python
X_train = X_train.drop(['Stops'],axis=1)
X_train = X_train.drop(['Trujet'],axis=1)
X_train = X_train.drop(['DEL'],axis=1)
X_train = X_train.drop(['DEL '],axis=1)
X_train = X_train.drop(['None'],axis=1)
X_train = X_train.drop(['TRV '],axis=1)
X_train = X_train.drop(['UDR '],axis=1)
X_train = X_train.drop(['VGA '],axis=1)
```

```python
X_test = X_test.drop(['Stops'],axis=1)
X_test = X_test.drop(['Trujet'],axis=1)
X_test = X_test.drop(['DEL'],axis=1)
X_test = X_test.drop(['DEL '],axis=1)
X_test = X_test.drop(['None'],axis=1)
X_test = X_test.drop(['TRV '],axis=1)
X_test = X_test.drop(['UDR '],axis=1)
X_test = X_test.drop(['VGA '],axis=1)
```

# Random Forest Regressor

```python
from sklearn.model_selection import RandomizedSearchCV
#Randomized Search CV

n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 50, cv = 5, verbose=2, random_state=4)
```

```python
rf_random.fit(X_train,y_train)
```

```python
y_pred=rf_random.predict(X_test)
```

```python
import seaborn as sns

sns.distplot(y_test-y_pred)
```

```python
plt.scatter(y_test,y_pred)
```

```python
from sklearn import metrics
print('MAE: ', metrics.mean_absolute_error(y_test,y_pred))
print('MSE: ', metrics.mean_squared_error(y_test,y_pred))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
